Example usigin urllib3 library to read data from an external URL, process them and convert them into a python data frame, before saving it in a CSV file. 

```python
def downloadFromURL(url, filename, mainpath, sep, delimiter, encoding):
    # firts import the library & make the request.
    import urllib3
    http = urllib3.PoolManager()
    r = http.request('GET', url)
    print("Status: %d" % (r.status))
    response = r.data

    # The response object contains a binary string, we've to convert it into a decoded string in UTF-8
    str_data = response.decode(encoding)

    # Split the array in every new row \n
    lines = str_data.split(delimiter)

    # The header is in the first line, so, extract it
    col_names= lines[0].split(sep)
    n_cols = len(col_names)

    # Generate a new dictionary that will contain the processes date from the URl
    counter = 0
    main_dict = {}
    for col in col_names:
        main_dict[col] = []

    # Process row by row to fill the dictionary
    for line in lines:
        # Skip the header (first row)
        if ( counter > 0 ):
            # Split every string by comas
            values = line.strip().split(sep)
            # Add the value to it's respective column
            for i in range(len(col_names)):
                main_dict[col_names[i]].append(values[i])
        counter += 1

    print("Rows in data set: %d & columns: %d" % (counter, n_cols))

    # Convert the dictionary into a Data set, and test the results
    df = pd.DataFrame(main_dict)
    # print(medals_df.head())

    # Choose where to store the new Data set
    fullpath = os.path.join(mainpath, filename)

    # Save it as CSV, JSON or Excel file
    df.to_csv(fullpath + ".csv")
    df.to_json(fullpath + ".json")
    df.to_excel(fullpath + ".xlsx")
    print("Files stored in: " + fullpath)
    return df
```
